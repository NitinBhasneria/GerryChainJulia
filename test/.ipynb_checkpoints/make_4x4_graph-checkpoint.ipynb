```python
# simple script to create a 4x4 graph to test the population constraint
import matplotlib.pyplot as plt
import networkx as nx
from gerrychain.graph import Graph

graph = nx.grid_graph([4,4])
graph = nx.convert_node_labels_to_integers(graph)
# node_coords = list(graph.nodes())

populate(graph)
assign(graph)

# we are going to use gerrychain's to_json method to convert the graph
g = Graph(graph)
g.to_json("../julia_flips/test_grid_4x4.json")

# color by assignment
plt.figure()
nx.draw(
    graph,
#     pos={x: x for x in node_coords},
    node_color=[graph.nodes[x]["assignment"] for x in graph.nodes()]
)
plt.show()
```

```python
def populate(graph):
    # calculatedly populate the graph
    graph.nodes[0]["population"] = 20
    graph.nodes[6]["population"] = 20
    graph.nodes[8]["population"] = 20
    graph.nodes[15]["population"] = 20
    
    graph.nodes[1]["population"] = 10
    graph.nodes[2]["population"] = 10
    graph.nodes[3]["population"] = 10
    graph.nodes[5]["population"] = 10
    graph.nodes[9]["population"] = 10
    graph.nodes[10]["population"] = 10
    graph.nodes[11]["population"] = 10
    graph.nodes[13]["population"] = 10
    
    graph.nodes[4]["population"] = 1
    graph.nodes[7]["population"] = 1
    graph.nodes[12]["population"] = 1
    graph.nodes[14]["population"] = 1
    
def assign(graph):
    # assignments
    graph.nodes[0]["assignment"] = 1
    graph.nodes[1]["assignment"] = 1
    graph.nodes[4]["assignment"] = 1
    graph.nodes[5]["assignment"] = 1

    graph.nodes[2]["assignment"] = 2
    graph.nodes[3]["assignment"] = 2
    graph.nodes[6]["assignment"] = 2
    graph.nodes[7]["assignment"] = 2

    graph.nodes[8]["assignment"] = 3
    graph.nodes[9]["assignment"] = 3
    graph.nodes[12]["assignment"] = 3
    graph.nodes[13]["assignment"] = 3

    graph.nodes[10]["assignment"] = 4
    graph.nodes[11]["assignment"] = 4
    graph.nodes[14]["assignment"] = 4
    graph.nodes[15]["assignment"] = 4
```
